```python
import numpy as np
import pandas as pd
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')

file1.head()
```

```python
#unify the columns names before concating them.

file1=file1.rename(columns={'ST':'State','GENDER':'Gender'})
file2=file2.rename(columns={'ST':'State','GENDER':'Gender'})
file3=file3.rename(columns={'ST':'State','GENDER':'Gender'})

file2
```

```python
# Aggregate the data into one Data Frame using Pandas. 
file4=pd.concat([file1,file2,file3], axis=0)
file4
```

```python
#Standardizing header names

def lower_case_column_names(file4):
       #list comprenhension 
    file4.columns=[header.lower() for header in file4.columns]
        
    return file4

```

```python
#Working with data types – Check the data types of all the columns and fix the incorrect ones (for ex. customer lifetime value and number of open complaints ). Hint: remove the percentage from the customer lifetime value and truncate it to an integer value.

file4.info() 
```
