```python
# 1. Aggregate the data into one Data Frame using Pandas. 

    # 1.1 Importing the data frames using libraries 
    
import numpy as np
import pandas as pd
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')

```

```python
  # 1.2 Unify the columns names before concating them.

file1=file1.rename(columns={'ST':'State','GENDER':'Gender'})
file2=file2.rename(columns={'ST':'State','GENDER':'Gender'})
file3=file3.rename(columns={'ST':'State','GENDER':'Gender'})


```

```python
# 1.3 Aggregate the data into one Data Frame using Pandas.
    
file4=pd.concat([file1,file2,file3], axis=0)

```

```python
#2. Standardizing header names

def lower_case_column_names(file4):
       #list comprenhension 
    file4.columns=[header.lower() for header in file4.columns]
        
    return file4

file4=lower_case_column_names(file4)
file4
```

```python
# 3. Removing duplicates

file4.drop_duplicates()
```

```python
# 4. Deleting and rearranging columns â€“ delete the column customer as it is only a unique identifier for each row of data

def drop_columns(file4):
    file4.drop(columns=['customer'], inplace=True)
    return file4

drop_columns(file4)
```

```python
# 4. Working with data types â€“ 
    # Check the data types of all the columns and fix the incorrect ones (for ex. customer lifetime value and number of open complaints ). Hint: remove the percentage from the customer lifetime value and truncate it to an integer value.

file4.info() 
```

```python
#how to find the data type and quantity of data in a column.
print(file4['customer lifetime value'].apply(type).value_counts())
```

```python
# 5. Remove the % of customer lifetime value

file4['customer lifetime value']=file4['customer lifetime value'].apply(lambda x: x.replace('%', '') if isinstance(x, str) else x)
```

```python
# Showing that the symbol"%" is removed. 

file4
```

```python
#Reset index in the data frame to avoid multiple "o" index

file4=file4.reset_index(drop=True)
```

```python

file4['customer lifetime value'] = file4['customer lifetime value'].astype(float)
```

```python
# Calculation of the mean to replace NAN with this value 

mean_clv = file4['customer lifetime value'].mean()
mean_clv
```

```python
# Replace NAN with the mean 
file4['customer lifetime value'] = file4['customer lifetime value'].fillna(mean_clv)
```

```python
file4
```

```python
#rounding the number to decimals

file4['customer lifetime value'] = file4['customer lifetime value'].round(0)
file4
```

```python
#extract the middle number of 'number of open complaints' which is changing between records.

file4['number of open complaints']=file4['number of open complaints'].apply(lambda x: x.replace('%', '') if isinstance(x, str) else x)
```
